```python
import pandas as pd 
from kiwipiepy import Kiwi
from FINAL.module.jh import JH
import numpy as np
import re
```

```python
kiwi = Kiwi(model_type="sbg", num_workers= 10)

########################################################################
kiwi.add_user_word("건", "XPN", 3) 
kiwi.add_user_word("건식", "NNG", 3)
kiwi.add_user_word("생", "XPN", -3)
kiwi.add_user_word("라떼", "NNG", -3)
kiwi.add_user_word("생크림", "NNG", -3)
kiwi.add_user_word("생선포", "NNG", 0)
kiwi.add_pre_analyzed_word("건포도", ["건/XPN", "포도/NNG"], 15)
kiwi.add_pre_analyzed_word("건호두", ["건/XPN", "호두/NNG"], 15)
kiwi.add_pre_analyzed_word("건담", ["건담/NNP"], 15)
kiwi.add_pre_analyzed_word("건망고", ["건/XPN", "망고/NNG"], 15)
kiwi.add_pre_analyzed_word("건조", ["건/XPN"], 15)
######################################################################### 
kiwi.add_pre_analyzed_word("생초코라떼", ["생/XPN", "초코/NNG","라떼/NNG"], 15)
kiwi.add_pre_analyzed_word("생맥주", ["생맥주/NNG"], 15)
kiwi.add_pre_analyzed_word("생맥", ["생맥주/NNG"], 15)
kiwi.add_pre_analyzed_word("생크림", ["생크림/NNG"], 15)
kiwi.add_pre_analyzed_word("생선포", ["생선포/NNG"], 15)
########################################################################
kiwi.add_pre_analyzed_word("계란", ["달걀/NNG"], 15)
```

```python
def replace_prefix_fresh(txt) :
    result = kiwi.analyze(txt)[0][0]
    for i in result : 
        if i.form == "생" :
            pattern = re.compile(i.form)
            return re.sub(pattern, f"({i.form})", str(txt))
        else :
            return str(txt)
```
